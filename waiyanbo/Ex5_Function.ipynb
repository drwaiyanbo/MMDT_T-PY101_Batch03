Ex 3: Write a function called number_of_factors that takes an integer and returns how many
factors the number has.

```python
# define funciton for factor
def number_of_factors (n):
  count = 0

  # check every number from 1 up to n
  for i in range(1, n+1):
    # use the modulo operator (%) to check for a remainder
    if n % i == 0:
     count = count + 1

  return count

# print out function
print(number_of_factors(20))
```

Ex4: Write a function called first_diff that is given two strings and returns the first location in
which the strings differ. If the strings are identical, it should return -1.Ignore the case in your function. It means h and H are the same.

```python
def first_diff(string1, string2):
  #convert strings into lowercase
  s1 = string1.lower()
  s2 = string2.lower()

  min_len = min(len(s1), len(s2))

  for i in range(min_len):
    if s1[i] != s2[i]:
      return i


  if len(s1) != len(s2):
    return min_len
  else:
    return -1

# test
print(first_diff("Hello", "hello"))
print(first_diff("Hello", "Hellos"))
```

Ex6: Write a function that solves the system of equations ax + by = e and cx +dy = f. Your function should takes 6 parameters, (a, b, c, d, e,f) and return the values of x and y.

```python
def solve_parameters(a, b, c, d, e, f):
  # value of x
  x = (e*d - b*f) / a
  # value of y
  y = (a*f - e*c) / b
  return x, y

print(solve_parameters(1, 2, 3, 4, 5, 6))
```

Ex7:  Write a function called change_case that given a string, returns a string with each upper  case letter replaced by a lower case letter and vice-versa

```python
def change_case(text):
  result = ""

  for char in text:
    if char.isupper():
      # if it's upper case, change to lower
      result += char.lower()
    elif char.islower():
      # if it's lower case, change to upper
      result += char.upper()
    else:
      # if it's not a letter, keep the same
      result += char
  return result

print(change_case("Hello World"))
```

Ex8:  Write a function called primes that is given a number n and returns a list of the first n  primes. Let the default value of n be 100.

```python
def primes(n=100):
  prime_list = []
  num = 2

  while len(prime_list) < n:
    is_prime = True

    for i in range(2, num):
      if num % i == 0:
        is_prime = False
        break

    if is_prime:
      prime_list.append(num)

    num += 1

  return prime_list

print(primes(5))
print(primes(6))
```

Ex9: A Tic-tac-toe board can be represented be a 3×3 two-dimensional list, where zeroes stand for  empty cells, ones stand for X’s and twos stand for O’s. Write a function that is given such a list and randomly chooses a spot in which to place  a 2 (O). The spot chosen must currently be empty and a spot must be chosen.

```python
import random

def place_o(board):
    # Find all coordinates (row, col) where the cell is empty (0)
    empty_spots = [(r, c) for r in range(3) for c in range(3) if board[r][c] == 0]

    # If there are empty spots, choose one randomly and place a 2
    if empty_spots:
        r, c = random.choice(empty_spots)
        board[r][c] = 2

# Test with a sample board
game_board = [[1, 2, 0], [0, 1, 0], [2, 2, 1]]
place_o(game_board)

for row in game_board:
    print(row)
```

Ex10:  Write a function that is given a 9 × 9 potentially solved Sudoku and returns True if it is solved correctly and False if there is a mistake. The Sudoku is correctly solved if there are
 no repeated numbers in any row or any column or in any of the nine “blocks.”


```python
def soduku_solved(board):
  def valid_unit(unit):
    # filter out zero
    nums = [n for n in unit if n !=0]
    return len(nums) == 9 and len(set(nums)) == 9


  # check rows
  for row in board:
    if not valid_unit(row):
      return False

  # check columns
  for col in zip(*board):
    if not valid_unit(col):
      return False

  # check blocks
  for r in range(0, 9, 3):
    for c in range(0, 9, 3):
      block = []
      for i in range(3):
        for j in range(3):
          block.append(board[r+i][c+j])
      if not valid_unit(block):
        return False

  return True
```
